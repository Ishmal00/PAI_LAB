```python
# Testing the integration of Transformer Engine and NLP Pipeline
from transformer_engine import TransformerProcessor
from meeting_assistant import MeetingAssistant

# Initialize the assistant (This will load the BART and BERT models)
assistant = MeetingAssistant()
print("✅ Assistant and Transformer Models Loaded Successfully")
```

```python


# Sample transcript to test the logic
test_transcript = """
Ishmal: Welcome to the AI project sync. We need to finalize the UI by Friday.
Alex: I will handle the CSS and the download button.
Ishmal: That sounds good. We decided to use Flask for the backend.
Alex: Are we including a database?
Ishmal: No, we will stick to session history for now. This is an urgent priority.
"""

# Run analysis
insights = assistant.analyze_meeting(test_transcript)

print(f"--- AI Analysis Results ---")
print(f"Summary: {insights.summary}")
print(f"Participants Found: {insights.participants}")
print(f"Sentiment: {insights.sentiment}")
```

```python
# Manual Evaluation of Extraction Logic
def evaluate_logic():
    expected_tasks = 1
    detected_tasks = len(insights.action_items)
    
    accuracy = (detected_tasks / expected_tasks) * 100 if expected_tasks > 0 else 100
    print(f"Task Extraction Accuracy: {accuracy}%")
    print(f"Estimated Duration: {insights.duration_estimate}")

evaluate_logic()
```
